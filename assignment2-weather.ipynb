## Problem Statement
Create a jupyter notebook called assignment2-weather.ipynb that has a nice plot of the temperature over time 
("dryBulbTemperature_Celsius").  

## Assumptions
There is no field directly representing the time. We will use the *localReportEndDateTime* field.  

```python
# Import the required packages
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the source directory and filename
FILENAME= "weatherreadings1.csv"
DATADIR = "./"
```

## Summary Statistics
Following is a set of statistics summarizing the weatherreadings1.csv file.
We are most interested in the count, i.e. the number of entries to be plotted. 

```python
# Read the data into a PANDAS dataframe
df = pandas.read_csv(DATADIR + FILENAME)
df.describe()
```

## Temperature Plot
Following is the code associated with drawing the plot, and the resulting plot itself.

```python
fig, ax = plt.subplots()

x = df.localReportEndDateTime
y = df.dryBulbTemperature_Celsius

# Split out date and time - commented out for the time being
#index=0
#date1 = []
#time1 = []
#for index in range(len(x)):
    # the last character is not included (10, 19 in this case)
    #date1.append(x[index][0:10])
    #time1.append(x[index][11:19])
    #print ("date is", date1[index])
    #print ("time is", time1[index])
    #index+=1

ax.plot(x, y)

ax.set(xlabel = "Date/Time", ylabel = "Temperature", title = "Weather Readings")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

ax.set_xticks(x[::4])
ax.set_xticklabels(x[::4], fontsize=7)

# Pad margins so markers aren't clipped by the axes and tweak spacing so tick-labels aren't clipped
plt.margins(None, 0.2)
#plt.subplots_adjust(bottom=0.15)

plt.xticks(x, rotation='vertical')
plt.yticks(fontsize=7)
plt.grid(axis='x')
plt.show()
```
